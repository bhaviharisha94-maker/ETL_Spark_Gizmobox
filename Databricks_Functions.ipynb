```sql
create or replace function gizmobox.default.get_fullname(firstname string, surname string)
returns string
return concat(initcap(firstname), ' ', initcap(surname))
```

```sql
create or replace temp view tv_order_items as
select * from 
values(1,array('smartphone','laptop','monitor')),
(2,array('tablet','headphone','smartwatch')),
(3,array('keyboard','mouse'))
as orders(order_id, items);
```

```sql
select order_id,transform(items,x->upper(x)) UPPER_ITEMS from tv_order_items;
select order_id, filter(items, x-> x like '%smart%') fitler_ones from tv_order_items;
select order_id, exists(items, x-> x='monitor') as exits_check from tv_order_items;
```

```sql
create or replace temp view tv_order_items as
select * from 
values(1,array(
    named_struct('name','smartphone','price',690),
    named_struct('name','laptop','price',1190),
    named_struct('name','monitor','price',390))),
(2,array(named_struct('name','tablet','price',590),
    named_struct('name','headphone','price',190),
    named_struct('name','smartwatch','price',290))),
(3,array(named_struct('name','keyboard','price',80),
    named_struct('name','mouse','price',50)))
as orders(order_id, items);
```

```sql
select order_id,
transform(items, x -> named_struct(
    'name',upper(x.name),
    'price',round(x.price*1.1,0)
))upperitem_with_tax from tv_order_items;
```

```sql
select order_id,
aggregate(items,0,(acc,x) -> acc+x.price) as TOtal_price
from tv_order_items
```

```sql
create or replace temp view order_item_price as
select * from values
    (1, map('smartphone',690,'laptop',1190,'monitor',390)),
    (2, map('tablet',590,'headphone',190,'smartwatch',290)),
    (3, map('keyboard',80,'mouse',50))
as orders(order_id, item_prices);
```

```sql
select order_id,
    transform_keys(item_prices,(k,v) -> upper(k)) as UPPER_ITEMS,
    transform_values(item_prices,(k,v) -> round(v*1.1,0)) as PRICE_WITH_TAX,
    map_filter(item_prices,(k,v) -> v>500) as PREMIUM_ITEM
     from order_item_price;
```
